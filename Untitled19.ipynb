```python
infected5 = pd.read_csv("/content/infected_dataINDEX.csv") 

infected5.head()
```

```python
infected5["Date"] = pd.to_datetime(infected5['Date'], dayfirst=True)
infected5["Date"].dtype
```

# Prophet_maroc

```python
mycolumns_Morocco = ['Date','Morocco'] 
```

```python
infected_Morocco = infected5[mycolumns_Morocco]
```

```python
infected_Morocco
```

```python
infected_Morocco['Date'] = pd.to_datetime(infected_Morocco['Date'])
infected_Morocco.columns = ['ds', 'y']
```

```python
infected_Morocco
```

```python
model_Morocco = Prophet()
model_Morocco.fit(infected_Morocco)
```

```python
future_Morocco = model_Morocco.make_future_dataframe(periods=365 * 10)
forecast_Morocco = model_Morocco.predict(future_Morocco)
```

```python
forecast_Morocco
```
